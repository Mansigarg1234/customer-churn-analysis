```python
# Assuming 'df' is your DataFrame

# Create a crosstab to count occurrences and calculate percentages
cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plotting
ax = cross_tab.plot(kind='bar', stacked=True, figsize=(5, 5), color=['blue', 'orange'])

# Adding percentage labels
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn Distribution by Senior Citizen Status')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.show()
```

```python
# List of columns to plot (excluding 'customerID')
columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
           'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
           'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges',
           'TotalCharges', 'Churn']

# Number of plots
num_plots = len(columns)

# Set up the matplotlib figure, dynamically based on the number of plots
rows = (num_plots + 3) // 4  # Calculate rows needed for 4 columns per row
fig, axes = plt.subplots(nrows=rows, ncols=4, figsize=(20, 5 * rows))
axes = axes.flatten()

# Loop through the columns and create a count plot for each
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Turn off any unused subplots
for j in range(num_plots, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
